#### é€šè¿‡WebSocketæœåŠ¡åˆ†å‘æµå¼è¾“å‡ºç»“æœ

ä¸‹é¢å†æä¾›ä¸€ä¸ªç®€å•ç¤ºä¾‹ï¼Œæ¼”ç¤ºå¦‚ä½•å°†æµå¼æ•°æ®é€šè¿‡WebSocketè½¬å‘è¾“å‡ºï¼š

````python
"""æœ¬ç¤ºä¾‹é€šè¿‡tornadoè¿›è¡ŒWebSocketæœåŠ¡"""
"""æ­¤æ®µä¸ºæœåŠ¡ç«¯ä»£ç """
import Agently
import asyncio
import tornado.ioloop
import tornado.web
import tornado.websocket
import json

agent_factory = (
    Agently.AgentFactory()
        .set_settings("current_model", "ZhipuAI")
        .set_settings("model.ZhipuAI.auth", { "api_key": "" })
)

agent = agent_factory.create_agent()

class WebSocketHandler(tornado.websocket.WebSocketHandler):
    def on_message(self, message):
        try:
            data = json.loads(message)
            print("Data Received:", data)
            result = (
                agent
                    # ä»ä¼ é€’è¿‡æ¥çš„æ¶ˆæ¯ä¸­è¯»å–"input"å­—æ®µçš„å€¼ï¼Œä¼ å…¥.input()
                    .input(data["input"])
                    # å°†æµå¼è¾“å‡ºç»“æœç”¨{ "event": "delta", "data": <è¾“å‡ºå—å†…å®¹> }çš„æ ¼å¼è¿”å›
                    .on_delta(lambda data: self.write_message(
                        json.dumps({
                            "status": 200,
                            "event": "delta",
                            "data": data,
                        })
                    ))
                    .start()
            )
            # åœ¨å…¨éƒ¨ç”Ÿæˆç»“æŸåï¼Œç”¨{ "event": "done", "data": <å®Œæ•´ç»“æœ> }çš„æ ¼å¼è¿”å›
            self.write_message(
                json.dumps({
                    "status": 200,
                    "event": "done",
                    "data": result,
                })
            )
        except Exception as e:
            print(e)
            self.write_message(json.dumps({
                "status": 400,
                "msg": str(e),
            }))

class Application(tornado.web.Application):
    def __init__(self):
        handlers = [
            ("/app", WebSocketHandler),
        ]
        super(Application, self).__init__(handlers)

app = Application()
app.listen(1565)
print("WebSocket æœåŠ¡å·²å¯åŠ¨")
tornado.ioloop.IOLoop.current().start()
````

#### ç»“åˆæ¡†æ¶å…¶ä»–èƒ½åŠ›ï¼Œæ„å»ºæ›´å¤æ‚çš„å¤šè½®å¯¹è¯

å½“ç„¶ï¼Œåœ¨ä¸Šé¢çš„åŸºç¡€å¯¹è¯æ–¹æ¡ˆä¹‹ä¸Šï¼Œæˆ‘ä»¬å¯ä»¥ç»“åˆæ¡†æ¶çš„å…¶ä»–èƒ½åŠ›ï¼Œæ„å»ºæ›´åŠ å¤æ‚çš„å¤šè½®å¯¹è¯ï¼Œä¾‹å¦‚ï¼šä½¿ç”¨Segmentæ–¹æ³•æŒ‡å¯¼Agentå›å¤çš„æ€è€ƒé€»è¾‘ã€‚

````python
def print_streaming_content(data: str):
    print(data, end="")

agent.set_settings("is_debug", False)
agent.active_session()
while True:
    user_input = input("[ç”¨æˆ·]: ")
    if user_input == "#exit":
        print("å†è§~ğŸ‘‹")
        break
    print("[åŠ©ç†]: ", end="")
    result = (
        agent
            .input(user_input)
            .segment(
                "thinking_tips",
                "ç”Ÿæˆä¸€æ®µæ–‡å­—ï¼Œç”¨é€‚å½“çš„æ€åº¦è¡¨è¾¾å¯¹ç”¨æˆ·è¾“å…¥ä¿¡æ¯çš„å…³æ³¨ï¼Œä½ æ­£åœ¨æ€è€ƒå¦‚ä½•å›å¤ï¼Œç”¨æ¢è¡Œç¬¦ç»“å°¾",
                print_streaming_content,
                is_streaming = True,
            )
            .segment(
                "required_info_list",
                [
                    {
                        "çŸ¥è¯†å¯¹è±¡": ("str", "å›ç­”{input}é—®é¢˜æ—¶ï¼Œéœ€è¦äº†è§£ç›¸å…³çŸ¥è¯†çš„å…·ä½“å¯¹è±¡"),
                        "å·²çŸ¥ä¿¡æ¯": ("str", "æ ¹æ®ä¹‹å‰æ‰€æœ‰å¯¹è¯å†å²ï¼Œæ€»ç»“å·²çŸ¥ä¿¡æ¯"),
                        "æ˜¯å¦å®Œå¤‡": ("bool", "åˆ¤æ–­ä½ æ˜¯å¦ç¡®ä¿¡è‡ªå·±æ‹¥æœ‰{çŸ¥è¯†å¯¹è±¡}çš„å…³é”®çŸ¥è¯†æˆ–ä¿¡æ¯ï¼Œå¦‚æœä¸äº†è§£ï¼Œè¾“å‡ºfalse"),
                        "å…³é”®çŸ¥è¯†ç‚¹æˆ–ä¿¡æ¯è¡¥å……": ("str", "å¦‚æœ{æ˜¯å¦å®Œå¤‡}==falseï¼Œç»™å‡ºéœ€è¦äº†è§£çš„å…³é”®çŸ¥è¯†æˆ–éœ€è¦ç”¨æˆ·æä¾›çš„ä¿¡æ¯è¡¥å……ï¼Œå¦åˆ™è¾“å‡ºç©ºå­—ç¬¦ä¸²''"),
                    }
                ],
            )
            .segment(
                "certain_reply",
                "æ ¹æ®{required_info_list}ç»™å‡ºå›å¤ï¼Œå±•å¼€è¯¦ç»†é™ˆè¿°è‡ªå·±äº†è§£çš„å…³é”®çŸ¥è¯†ç‚¹å†…å®¹ï¼Œç”¨æ¢è¡Œç¬¦ç»“å°¾ï¼Œå¦‚æœæ²¡æœ‰åˆ™è¾“å‡ºç©ºå­—ç¬¦ä¸²''",
                print_streaming_content,
                is_streaming = True,
            )
            .segment(
                "uncertain_reply",
                "æ ¹æ®{required_info_list}çš„ä¿¡æ¯ï¼Œå‘ç”¨æˆ·è¯´æ˜è‡ªå·±ä¸äº†è§£çš„ä¿¡æ¯ï¼Œè¯·ç”¨æˆ·æä¾›æˆ–è‡ªè¡ŒæŸ¥æ‰¾ï¼Œç”¨æ¢è¡Œç¬¦ç»“å°¾ï¼Œå¦‚æœæ²¡æœ‰åˆ™è¾“å‡ºç©ºå­—ç¬¦ä¸²''",
                print_streaming_content,
                is_streaming = True,
            )
            .segment(
                "next_topic_suggestions",
                "æ ¹æ®ä¹‹å‰æ‰€æœ‰ç”Ÿæˆå†…å®¹çš„ä¿¡æ¯ï¼Œç»™å‡ºæ¥ä¸‹æ¥å¯ä»¥è¿›ä¸€æ­¥è®¨è®ºçš„é—®é¢˜æˆ–è¯é¢˜å»ºè®®ï¼Œå¦‚æœæ²¡æœ‰åˆ™è¾“å‡ºç©ºå­—ç¬¦ä¸²''",
                print_streaming_content,
                is_streaming = True,
            )
            .start()
    )
    print("")
````

å½“ç„¶ï¼Œåœ¨æ›´åŠ å¤æ‚çš„åœºæ™¯ä¸­ï¼Œæˆ‘ä»¬éœ€è¦è‡ªå·±å®šåˆ¶ç®¡ç†å¯¹è¯è®°å½•ï¼Œè€Œä¸æ˜¯ä½¿ç”¨è‡ªåŠ¨è®°å½•æ–¹æ¡ˆï¼ŒAgentlyæ¡†æ¶ä¹Ÿä¸ºæ‚¨æ‰‹åŠ¨ç®¡ç†å¯¹è¯è®°å½•æä¾›äº†ä¾¿æ·çš„æ–¹æ³•æ”¯æŒã€‚

##### å·¥å…·çš„å„ä¸ªå…³é”®ä¿¡æ¯éƒ¨åˆ†

è¦æ³¨å†Œå·¥å…·ï¼Œé¦–å…ˆè¦äº†è§£å·¥å…·ç”±å“ªäº›å…³é”®ä¿¡æ¯ç»„æˆï¼Œä»¥åŠè¿™äº›å…³é”®ä¿¡æ¯åœ¨Agentå®ä¾‹çš„å·¥ä½œè¿‡ç¨‹ä¸­çš„ä½œç”¨ï¼š

* **`tool_name`ï¼ˆå·¥å…·åç§°ï¼‰**ï¼šå¸®åŠ©Agentå®ä¾‹ç¡®å®šéœ€è¦è°ƒç”¨çš„å·¥å…·çš„åç§°ï¼Œå¯ä»¥è§†ä½œå·¥å…·çš„IDï¼Œåœ¨å½“å‰Agentå®ä¾‹å¯ç”¨çš„å·¥å…·é›†ä¸­ä¸å¯é‡å¤ï¼›
* **`desc`ï¼ˆå·¥å…·æè¿°ï¼‰**ï¼šå¸®åŠ©Agentå®ä¾‹è¿›ä¸€æ­¥ç†è§£å·¥å…·å¯ä»¥è§£å†³çš„å…·ä½“é—®é¢˜ã€åœºæ™¯ã€ä½¿ç”¨é™åˆ¶ç­‰ï¼›
* **`args`ï¼ˆå‚æ•°å­—å…¸ï¼‰**ï¼šç”¨å­—å…¸ï¼ˆdictï¼‰çš„æ ¼å¼å¸®åŠ©Agentå®ä¾‹ç†è§£è°ƒç”¨å·¥å…·éœ€è¦ä¼ å…¥çš„å‚æ•°è¦æ±‚ã€æ ¼å¼é™åˆ¶åŠå…¶å®ƒè¯´æ˜ï¼Œåœ¨è¿™é‡Œå¯ä»¥ä½¿ç”¨[ç¬¬ä¸‰ç« æåˆ°çš„Outputè¾“å‡ºç»“æ„è¡¨è¾¾è¯­æ³•](#scrollTo=cFi7tB1rE3uK)ï¼›
* **`func`ï¼ˆå·¥å…·å‡½æ•°ï¼‰**ï¼šéœ€è¦è¢«è°ƒç”¨çš„å·¥å…·å‡½æ•°æœ¬ä½“ï¼Œå½“Agentå®ä¾‹æŒ‰è¦æ±‚ç”Ÿæˆå¹¶ä¼ é€’ç›¸å…³å‚æ•°åï¼Œèƒ½å¤Ÿä»è¿™ä¸ªå‡½æ•°è·å–åˆ°è°ƒç”¨ç»“æœå€¼ã€‚

åœ¨è¿™é‡Œï¼Œæˆ‘ä»¬å…ˆå‡†å¤‡ä¸€ä¸ªâ€œè·å–å½“å‰æ—¶é—´â€çš„ç®€å•å·¥å…·çš„ç›¸å…³ä¿¡æ¯ï¼š

````python
from datetime import datetime
import pytz
def get_current_datetime(timezone):
    tz = pytz.timezone(timezone)
    return datetime.now().astimezone(tz)

tool_info = {
    "tool_name": "get_now",
    "desc": "get current data and time",
    "args": {
        "timezone": (
            "str",
            "[*Required] Timezone string used in pytz.timezone() in Python"
        )
    },
    "func": get_current_datetime
}
````

##### å‘å…¬å…±å·¥å…·æ± æ³¨å†Œå…¬å…±å·¥å…·



å‘å…¬å…±å·¥å…·æ± æ³¨å†Œå·¥å…·æœ‰ä¸¤ç§æ–¹æ³•ï¼Œåœ¨å…¥é—¨ç¯‡ä¸­ï¼Œæˆ‘ä»¬å°†åªä»‹ç»ä¾›åº”ç”¨å¼€å‘è€…ç¾¤ä½“ä½¿ç”¨çš„**å‘å…¨å±€å·¥å…·ç®¡ç†å™¨ï¼ˆglobal_tool_managerï¼‰æ³¨å†Œå…¬å…±å·¥å…·ä¿¡æ¯**çš„æ–¹æ³•ã€‚åœ¨åç»­æ›´æ–°çš„é¢å‘æ¡†æ¶æ’ä»¶å¼€å‘è€…çš„ç›¸å…³æ–‡æ¡£ä¸­ï¼Œæˆ‘ä»¬å†ä»‹ç»å¦‚ä½•ä»¥æ’ä»¶çš„æ–¹å¼è¿›è¡Œå…¬å…±å·¥å…·çš„å¼€å‘ã€‚

##### å‘Agentå®ä¾‹æ³¨å†Œç§æœ‰å·¥å…·

å‘Agentå®ä¾‹æ³¨å†Œç§æœ‰å·¥å…·ä¹Ÿæœ‰ä¸¤ç§æ–¹æ³•ï¼Œè®©æˆ‘ä»¬å…ˆä»‹ç»**ä½¿ç”¨.register_tool()æŒ‡ä»¤å‘Agentå®ä¾‹æ³¨å†Œç§æœ‰å·¥å…·**çš„æ–¹æ³•ï¼š

````python
# åˆ›å»ºä¸€ä¸ªtool_using_agent
tool_using_agent = agent_factory.create_agent()

# é€šè¿‡.register_tool()çš„æŒ‡ä»¤æ³¨å†Œå·¥å…·ï¼Œå‚æ•°å’Œä¸Šä¸€ä¸ªæ¡ˆä¾‹çš„è¦æ±‚ä¸€è‡´
tool_using_agent.register_tool(
    tool_name = tool_info["tool_name"],
    desc = tool_info["desc"],
    args = tool_info["args"],
    func = tool_info["func"],
)

# åŒæ ·ä¹Ÿå¯ä»¥ä½¿ç”¨kwargsæ–¹å¼è¾“å…¥å‚æ•°
tool_using_agent.register_tool(**tool_info)
````

åŒæ—¶ï¼ŒAgentlyæ¡†æ¶è¿˜æä¾›äº†å¦ä¸€ç§æ–¹ä¾¿çš„Agentå®ä¾‹ç§æœ‰å·¥å…·æ³¨å†Œæ–¹æ³•ï¼š**é€šè¿‡Agentå®ä¾‹çš„agent.tool()è£…é¥°å™¨æ³¨å†Œç§æœ‰å·¥å…·**ï¼š

````python
from datetime import datetime
import pytz
# è¿™é‡Œæˆ‘ä»¬åˆ›å»ºå¦ä¸€ä¸ªä¸åŒåç§°çš„agent
tool_using_with_decorator_agent = agent_factory.create_agent()

# ä½ å¯ä»¥å°†é™¤äº†funcçš„tool_infoå­—å…¸çš„ä¿¡æ¯ä¼ å…¥è£…é¥°å™¨å‡½æ•°
# å½“æ²¡æœ‰ä¼ å…¥ä¿¡æ¯ç»™è£…é¥°å™¨å‡½æ•°çš„æ—¶å€™ï¼Œè£…é¥°å™¨å‡½æ•°å°†ä¼šä»è¢«è£…é¥°å‡½æ•°ä¸­å°è¯•å¯»æ‰¾ç›¸å…³ä¿¡æ¯
@tool_using_with_decorator_agent.tool(tool_name="get_now")
def get_current_datetime_decorated(
    # å¦‚æœæ²¡æœ‰ç»™å®štool_info["tool_name"]ï¼Œä¸Šé¢çš„å‡½æ•°åç§°ä¼šä½œä¸ºtool_nameè¢«ä¼ é€’
    # å‚æ•°å®šä¹‰éƒ¨åˆ†çš„ä¿¡æ¯ä¼šè¢«è£…é¥°å™¨è§£æåˆ°tool_info["args"]ä¸­
    timezone: (
        "str",
        "[*Required] Timezone string used in pytz.timezone() in Python"
    )
    # :å·ä¹‹åçš„ä¿¡æ¯ä¼šè¢«ä½œä¸ºå‚æ•°è¯´æ˜ä¼ é€’ï¼Œæ”¯æŒtupleæ–¹å¼å£°æ˜ç±»å‹+è¯¦ç»†è¯´æ˜
    # ä¹Ÿå¯ä»¥ä½¿ç”¨çº¯å­—ç¬¦ä¸²å½¢å¼è¿›è¡Œè¯´æ˜
):
    """get current data and time"""
    #é€šè¿‡ä¸Šé¢çš„docstringçš„æ–¹å¼ï¼Œå¯ä»¥ä¼ é€’tool_info["desc"]æ‰€éœ€çš„å·¥å…·æè¿°ä¿¡æ¯
    tz = pytz.timezone(timezone)
    return datetime.now().astimezone(tz)

# ä¸Šé¢è¿™ä¸ªç»“æ„ä¼ é€’çš„ä¿¡æ¯ä¸å‰ä¸€ä¸ªä¾‹å­ä¸­ä¼ é€’çš„å·¥å…·æ³¨å†Œä¿¡æ¯ä¸€è‡´
````

#### åœ¨Agentå®ä¾‹è¯·æ±‚æ—¶ä½¿ç”¨å·¥å…·

````python
# æˆ‘ä»¬ç”¨ä¸€ä¸ªä¸´æ—¶agentå®ä¾‹æ¥è°ƒç”¨å…¨å±€å·¥å…·

result = (
    agent_factory.create_agent()
        #å£°æ˜ä½¿ç”¨å…¨å±€å·¥å…·ä¸­æˆ‘ä»¬åˆšæ³¨å†Œçš„"get_now"
        .use_public_tools(["get_now"])
        # é€šè¿‡è¾“å…¥ç›¸å…³è¯¢é—®å†…å®¹æµ‹è¯•å·¥å…·è°ƒç”¨æ•ˆæœ
        .input("æˆ‘åœ¨åŒ—äº¬ï¼Œç°åœ¨å‡ ç‚¹äº†ï¼Ÿ")
        .start()
)
print(result)
````

````python
# åŒæ ·çš„ï¼Œä½¿ç”¨è£…é¥°å™¨æ³¨å†Œå·¥å…·çš„agentä¹Ÿä¸éœ€è¦é¢å¤–å£°æ˜ä½¿ç”¨å·¥å…·
result = tool_using_with_decorator_agent.input("æˆ‘åœ¨åŒ—äº¬ï¼Œç°åœ¨å‡ ç‚¹äº†ï¼Ÿ").start()
print(result)
````

#### ä½¿ç”¨.must_call()æ–¹æ³•è·å–å·¥å…·è°ƒç”¨å‚æ•°ï¼Œç”±å¼€å‘è€…è‡ªè¡Œè°ƒç”¨

åœ¨æŸäº›åœºæ™¯ä¸‹ï¼Œä½œä¸ºå¼€å‘è€…æˆ‘ä»¬å¸Œæœ›è‡ªè¡Œå®Œæˆå·¥å…·çš„è°ƒç”¨ä»¥åŠè¿”å›å€¼çš„åç»­å¤„ç†ç¼–æ’ï¼Œåªéœ€è¦Agentå®ä¾‹æ ¹æ®è¾“å…¥çš„ä¿¡æ¯å¸®åŠ©æˆ‘ä»¬ç”Ÿæˆå·¥å…·è°ƒç”¨æ‰€éœ€çš„å‚æ•°å³å¯ã€‚

é’ˆå¯¹è¿™ç§åœºæ™¯ï¼ŒAgentlyæ¡†æ¶æä¾›äº†`agent.must_call()`æ–¹æ³•ï¼Œä¸‹é¢æ˜¯å…·ä½“ç”¨æ³•ï¼š

````python
# é¦–å…ˆæˆ‘ä»¬éœ€è¦æŠŠæƒ³è¦ç”Ÿæˆè°ƒç”¨å‚æ•°çš„å·¥å…·å‡½æ•°æ³¨å†Œç»™agent
# åœ¨è¿™é‡Œæˆ‘ä»¬è¿˜æ˜¯ä½¿ç”¨ä¸Šé¢çš„è·å–æ—¶é—´å‡½æ•°

must_call_agent = agent_factory.create_agent()

@must_call_agent.tool(tool_name="get_now")
def get_current_datetime_decorated(
    timezone: (
        "str",
        "[*Required] Timezone string used in pytz.timezone() in Python"
    )
):
    """get current data and time"""
    tz = pytz.timezone(timezone)
    return datetime.now().astimezone(tz)

# ç„¶åï¼Œæˆ‘ä»¬ä½¿ç”¨ä¸‹é¢çš„æ–¹å¼å‘Šè¯‰agentä½ éœ€è¦è·å¾—æŒ‡å®šå·¥å…·å‡½æ•°çš„è°ƒç”¨å‚æ•°
result = (
    agent
        .must_call("get_now") #<-å¡«å…¥ä½ å¸Œæœ›ç”Ÿæˆè°ƒç”¨å‚æ•°çš„å·¥å…·å‡½æ•°å
        .input("æˆ‘åœ¨åŒ—äº¬ï¼Œç°åœ¨å‡ ç‚¹äº†ï¼Ÿ") #<-æä¾›å…¶ä»–ä¸agentå®ä¾‹äº¤äº’çš„ä¿¡æ¯ï¼Œå¯ä»¥ä¸ä»…é™äº.input()
        .start()
)
print(result)
````

åœ¨ä¸Šé¢æ¡ˆä¾‹çš„è¿”å›å€¼ä¸­ï¼Œä½ ä¼šå‘ç°ï¼Œè¿”å›å€¼é‡Œæœ‰ä¸‰ä¸ªå­—æ®µï¼Œè¿™ä¸‰ä¸ªå­—æ®µå°†å¯¹å¼€å‘è€…è¿›è¡Œåç»­å·¥å…·å‡½æ•°è°ƒç”¨æä¾›ä»¥ä¸‹å¸®åŠ©ï¼š

* **`can_call`**ï¼šå‘ŠçŸ¥å¼€å‘è€…å½“å‰è¾“å…¥çš„ä¿¡æ¯æ˜¯å¦è¶³å¤Ÿå®Œæˆå‡½æ•°çš„è°ƒç”¨ï¼Œå¦‚æœè¯¥é¡¹ä¸º`True`ï¼Œåˆ™`args`å­—æ®µä¸­æä¾›çš„ä¿¡æ¯å¯ä»¥è¢«ç”¨äºå‡½æ•°è°ƒç”¨ï¼›å¦‚æœè¯¥é¡¹ä¸º`False`åˆ™è¡¨ç¤ºå½“å‰è¾“å…¥çš„ä¿¡æ¯ä¸å¤Ÿå®Œå¤‡ï¼Œä¸åº”è¯¥ç»§ç»­è¿›è¡Œå‡½æ•°è°ƒç”¨ï¼›
* **`args`**ï¼šç¬¦åˆå‡½æ•°è°ƒç”¨æ‰€éœ€çš„å‚æ•°è¦æ±‚çš„å‚æ•°å­—å…¸ï¼ˆå¯è¢«ä»¥**kwargsé£æ ¼ä¼ é€’ç»™ç›®æ ‡å·¥å…·å‡½æ•°ï¼‰ï¼›
* **`question`**ï¼šå½“`can_call`ä¸º`False`æ—¶ï¼Œ`question`å­—æ®µå°†ä¸ºå¼€å‘è€…æä¾›ä¸€ä¸ªå»ºè®®é—®é¢˜ï¼Œå¸®åŠ©å¼€å‘è€…ç†è§£ç¼ºå¤±çš„ä¿¡æ¯æ˜¯ä»€ä¹ˆï¼Œå› ä¸ºè¿™ä¸ªé—®é¢˜æ˜¯é¢å‘ç”¨æˆ·è¾“å…¥ä¿¡æ¯çš„åé¦ˆï¼Œè¿™ä¸ªå»ºè®®é—®é¢˜ä¹Ÿå¯ä»¥ç›´æ¥ä¼ é€’ç»™ç”¨æˆ·ï¼Œç”¨äºæç¤ºç”¨æˆ·åº”è¯¥è¡¥å……å“ªäº›ä¿¡æ¯ã€‚
