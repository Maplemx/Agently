```python
import Agently
agent_factory = Agently.AgentFactory()

# æ¨¡å‹è®¾ç½®

## é€‰æ‹©ï¼šOpenAI - GPT 3.5-turbo | GPT 4 | GPT 4 Vision
agent_factory\
    .set_settings("current_model", "OpenAI")\
    .set_settings("model.OpenAI.auth", { "api_key": "" })
    ## æ›´æ¢Base URLï¼ˆå¯é€‰ï¼‰
    ## å¦‚æœä½ ä½¿ç”¨äº†å…¶ä»–æœåŠ¡ä¾›åº”å•†æä¾›çš„è½¬å‘æ¥å£
    ## æˆ–è€…ä½¿ç”¨äº†FastAPIç­‰éµå¾ªOpenAI Clientè§„èŒƒçš„æœ¬åœ°æ¨¡å‹æœåŠ¡ä»£ç†å·¥å…·ï¼Œéœ€è¦æ›´æ¢Base URL
    ## è¯·ä½¿ç”¨ä¸‹é¢çš„è®¾ç½®æ¥è®¾ç½®Base URL
    #.set_settings("model.OpenAI.url", "https://redirect-service-provider/api/v1")\
    ## è°ƒæ•´æ¨¡å‹è¯·æ±‚å‚æ•°optionsï¼ˆå¯é€‰ï¼‰
    ## å¦‚æœä½ éœ€è¦åˆ‡æ¢æ¨¡å‹ï¼Œæˆ–æ˜¯è°ƒæ•´è¯·æ±‚å‚æ•°ï¼Œå¯ä»¥ä½¿ç”¨ä¸‹é¢çš„è®¾ç½®è¿›è¡Œè°ƒæ•´
    ## å‚æ•°è¯´æ˜è¯·å‚è€ƒOpenAI Platformçš„å®˜æ–¹æ–‡æ¡£
    #.set_settings("model.OpenAI.options", { "model": "gpt-4" })\
    ## â—ï¸ä½¿ç”¨ä»£ç†ï¼ˆå¯é€‰ï¼‰
    ## å¦‚æœä½ åœ¨æœ¬åœ°æœºå™¨ä½¿ç”¨äº†æ¯”å¦‚Clashã€VPNã€V2Rayç­‰ä»£ç†å·¥å…·æ¥è®¿é—®OpenAIçš„API
    ## ä½ éœ€è¦æŸ¥çœ‹è¿™äº›ä»£ç†å·¥å…·æä¾›çš„ä»£ç†æœåŠ¡åœ°å€ï¼Œå¹¶å¡«å…¥ä¸‹é¢çš„è®¾ç½®
    ## æ³¨æ„ï¼šä½¿ç”¨Google Colabæ—¶ä¸€èˆ¬ä¸éœ€è¦è®¾ç½®ä»£ç†
    #.set_proxy("http://127.0.0.1:7890")

## é€‰æ‹©ï¼šæ™ºè°±AI - ChatGLM-turbo | CharacterGLM
## å¦‚ä½•è·å–ï¼šhttps://open.bigmodel.cn/overview
'''
agent_factory\
    .set_settings("current_model", "ZhipuAI")\
    .set_settings("model.ZhipuAI.auth", { "api_key": "" })
'''
## é€‰æ‹©ï¼šç™¾åº¦æ–‡å¿ƒå¤§æ¨¡å‹
## å¦‚ä½•è·å–ï¼šhttps://aistudio.baidu.com/index
'''
agent_factory\
    .set_settings("current_model", "ERNIE")\
    .set_settings("model.ERNIE.auth", {
        "aistudio": "",
    })
'''

# å¼€å§‹è¯•ç”¨
## ä½ å¯ä»¥ä½¿ç”¨ä¸‹é¢çš„æ–¹å¼æ‰“å¼€è°ƒè¯•æ¨¡å¼ï¼ˆdebug modeï¼‰ï¼Œæ‰“å°è¯·æ±‚æ•°æ®ï¼Œå¹¶è·å¾—å®æ—¶è¾“å‡ºåé¦ˆ
## æ³¨æ„ï¼šè°ƒè¯•æ¨¡å¼å’Œæ ·ä¾‹ä»£ç ä¸­çš„.on_delta(lambda data: print(data, end=""))
##      ä¼šäº§ç”Ÿæ˜¾ç¤ºå†²çªï¼Œæ‰“å¼€åè¯·æ‰‹åŠ¨åˆ é™¤.on_delta(lambda data: print(data, end=""))
#agent_factory.set_settings("is_debug", True)

## åœ¨è¿™é‡Œç¼–å†™ä½ çš„ä»£ç ï¼š
agent = agent_factory.create_agent()
agent.active_session()
while True:
    user_input = input("[ç”¨æˆ·]: ")
    if user_input == "#exit":
        break
    print("[AGENT]: ", end="")
    agent\
        .input(user_input)\
        .on_delta(lambda data: print(data, end=""))\
        .start()
    print("")
```
